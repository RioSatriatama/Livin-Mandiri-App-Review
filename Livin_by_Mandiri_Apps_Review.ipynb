# **Import Library**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

# **Import Dataset**

```python
df = pd.read_csv('/content/reviews.csv')
df
```

# **Create Ploting / Pie chart**

```python
fig, ax = plt.subplots(figsize = (5, 5))
colors = sns.color_palette('pastel')

count_rating = [count for count in df['rating'].value_counts()]
labels_rating = list(df['rating'].value_counts().index)

plt.title('Ratings', fontsize = 16)
ax.pie(x=count_rating, labels=labels_rating, autopct = '%1.1f%%',
       textprops={'fontsize': 14}, colors=colors)

plt.show
```

```python
text = " ".join(review for review in df[df.rating <= 5].review)
stopwords = ['nya', 'livin', 'yg', 'ada', 'aja', 'dan', 'saya',
             'ini', 'itu', 'apk', 'mandiri', 'di', 'aplikasi',
             'aplikasinya']
wordcloud = WordCloud(collocations = True, max_font_size = 50,
                      stopwords = stopwords, max_words = 100,
                      background_color = "white").generate(text)
fig, ax = plt.subplots(figsize = (10, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
